**Data Analysis**

Univariate Analysis: Analysing Single Values and Their Distribution

```python
#Price Distribution
data['price']
```

```python
#identify outliers
df = data[data['price'] < 2000]

sns.boxplot(data=df, x='price')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data = df, x = 'price', bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
df.dtypes
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data = df, x = 'availability_365')
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

Price per Neigborhood

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

**Feature Engineering**

```python
#creating a new column for price per bed
df['price per bed']=df['price']/df['beds']
df['price per bed']
```

```python
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
#price dependency on neighbourhood group
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
```

**Bi Variate Analysis**: One variable dependency in another variable

```python
#relationship price and reviews
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.show()
```

Plot all the number columns

```python
sns.scatterplot(data=df, x='longitude', y='latitude', hue='room_type')
plt.title('Geographical Distribution of AirBnB Listing')
plt.show()
```
